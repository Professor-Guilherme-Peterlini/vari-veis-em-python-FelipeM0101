````python
#atividade numero 7
itens: list = []
itens =[]
def adicionar(item):
    itens.append(item)
    print(f"Item - {item} adicionado.")

def remover_item(itemR):
    if itemR in itens:
        itens.remove(itemR)
        print(f"O Item - {itemR} foi removido.")
    else:
        print(f"Não o Item - {itemR}.")

def mostrar_lista():
    if itens:
        print("Lista de Compras")
        for i, item in enumerate(itens,start=1):
            print(f"{i} - {item}")
    else:
        print("Lista de compras vazia.")

while True:
    print("\n1. Adicionar item a lista")
    print("2. Remover item da lista")
    print("3. Ver lista compras")
    print("4. Sair")
    escolha = input("Escolha uma opcao (1/2/3/4): ")
    if escolha == "1":
        item = input("Digite um item a ser adicionado :")
        adicionar(item)
    elif escolha == "2":
        item = input("Digite o item a ser removido :")
        remover_item(item)
    elif escolha == "3":
       mostrar_lista()
    elif escolha == "4":
       print("Saindo do programa.")
       break
    else:
        print("Opção Invalida! Tente novamente.")
````

````python
#atividade numero 9
def bubble_sort(v):
    fim = len(v)
    while fim > 0:
        i = 0
        while i < fim - 1:
            if v[i] > v[i + 1]:
                v[i], v[i + 1] = v[i + 1], v[i]
            i += 1
        fim -= 1
    return v


listaNumeros = list()
for c in range(1, 6):
    listaNumeros.append(int(input(f'Digite o {c}º valor: ')))

print(bubble_sort(listaNumeros))
````

````python
#atividade 11
import random
import string

def gerar_senha(comprimento):
    caracteres = string.ascii_letters + string.digits + string.punctuation
    senha = ''.join(random.choice(caracteres) for _ in range(comprimento))
    return senha

def menu():
    while True:
        try:
            comprimento = int(input("Digite o comprimento da senha: "))
            if comprimento <= 0:
                print("Por favor, insira um número positivo.")
                continue
            senha = gerar_senha(comprimento)
            print(f"Sua senha gerada é: {senha}")
        except ValueError:
            print("Por favor, insira um número válido.")

        continuar = input("Deseja gerar outra senha? (s/n): ").strip().lower()
        if continuar != 's':
            print("Saindo do programa. Até logo!")
            break

# Executa o menu
menu()
````

````python
#atividade 12
def conversor_moedas(valor, moeda_origem, moeda_destino):
    # Taxas de câmbio fixas
    taxas = {
        'USD': 1.0,      # Dólar americano
        'EUR': 0.85,     # Euro
        'BRL': 5.25,     # Real brasileiro
        'JPY': 110.0     # Iene japonês
    }

    # Verifica se as moedas estão disponíveis
    if moeda_origem not in taxas or moeda_destino not in taxas:
        return None

    # Converte o valor da moeda de origem para dólares
    valor_em_usd = valor / taxas[moeda_origem]

    # Converte o valor de dólares para a moeda de destino
    valor_convertido = valor_em_usd * taxas[moeda_destino]

    return valor_convertido

def menu():
    print("Conversor de Moedas")
    print("Moedas disponíveis: USD, EUR, BRL, JPY")

    while True:
        try:
            valor = float(input("Digite o valor a ser convertido: "))
            moeda_origem = input("Digite a moeda de origem (USD, EUR, BRL, JPY): ").strip().upper()
            moeda_destino = input("Digite a moeda de destino (USD, EUR, BRL, JPY): ").strip().upper()

            resultado = conversor_moedas(valor, moeda_origem, moeda_destino)

            if resultado is not None:
                print(f"{valor} {moeda_origem} é igual a {resultado:.2f} {moeda_destino}.")
            else:
                print("Moeda de origem ou destino inválida.")

        except ValueError:
            print("Por favor, insira um valor numérico válido.")

        continuar = input("Deseja realizar outra conversão? (s/n): ").strip().lower()
        if continuar != 's':
            print("Saindo do programa. Até logo!")
            break

menu()
````

````python
#atividade 12
import random

def escolher_palavra():
    palavras = ["python", "programacao", "desenvolvimento", "jogo", "computador"]
    return random.choice(palavras)

def exibir_forca(tentativas):
    estagios = [
        """
           -----
           |   |
           |   O
           |  /|\\
           |  / \\
           |
        """,
        """
           -----
           |   |
           |   O
           |  /|\\
           |  /
           |
        """,
        """
           -----
           |   |
           |   O
           |  /|
           |
           |
        """,
        """
           -----
           |   |
           |   O
           |
           |
           |
        """,
        """
           -----
           |   |
           |
           |
           |
           |
        """,
        """
           -----
           |
           |
           |
           |
           |
        """,
    ]
    return estagios[tentativas]

def jogar():
    palavra = escolher_palavra()
    letras_adivinhadas = []
    tentativas = 6
    acertos = False

    print("Bem-vindo ao Jogo da Forca!")

    while not acertos and tentativas > 0:
        print(exibir_forca(tentativas))
        print("Palavra: ", end=" ")

        for letra in palavra:
            if letra in letras_adivinhadas:
                print(letra, end=" ")
            else:
                print("_", end=" ")

        print("\n")

        tentativa = input("Adivinhe uma letra: ").lower()

        if tentativa in letras_adivinhadas:
            print("Você já tentou essa letra. Tente outra.")
        elif tentativa in palavra:
            letras_adivinhadas.append(tentativa)
            print("Bom trabalho! A letra está na palavra.")
        else:
            letras_adivinhadas.append(tentativa)
            tentativas -= 1
            print("Ops! A letra não está na palavra.")

        if all(letra in letras_adivinhadas for letra in palavra):
            acertos = True

    if acertos:
        print(f"Parabéns! Você adivinhou a palavra: {palavra}")
    else:
        print(exibir_forca(tentativas))
        print(f"Você perdeu! A palavra era: {palavra}")

jogar()
````

````python
#atividade 13
def fatorial(n):

    if n < 0:
        return "Fatorial não definido para números negativos."
    # Caso base: fatorial de 0 é 1
    elif n == 0:
        return 1
    else:
        return n * fatorial(n - 1)

numero = int(input("Digite um número para calcular o fatorial: "))
resultado = fatorial(numero)
print(f"O fatorial de {numero} é: {resultado}")
````

````python
#atividade 14
def validar_cpf(cpf):

    cpf = ''.join(filter(str.isdigit, cpf))

    if len(cpf) != 11 or not cpf.isdigit():
        return False


    if cpf == cpf[0] * 11:
        return False


    def calcular_digito(cpf, peso):
        soma = sum(int(cpf[i]) * peso[i] for i in range(len(peso)))
        resto = soma % 11
        return 0 if resto < 2 else 11 - resto


    peso1 = list(range(10, 1, -1))
    peso2 = list(range(11, 2, -1))

    digito1 = calcular_digito(cpf[:9], peso1)
    digito2 = calcular_digito(cpf[:9] + str(digito1), peso2)


    return cpf[-2:] == f"{digito1}{digito2}"

cpf_input = input("Digite o CPF (apenas números): ")
if validar_cpf(cpf_input):
    print("CPF válido.")
else:
    print("CPF inválido.")
````
